```python
def tidy_date(date_str): 
    if date_str == 'Last Friday': 
        return pd.to_datetime('2018-12-14')
    elif 'New Year\'s Day' in date_str: 
        year = date_str[-4:]
        return pd.to_datetime('{}-01-01'.format(year))
    elif 'Christmas Eve' in date_str: 
        year = date_str[-4:]
        return pd.to_datetime('{}-12-24'.format(year))
    elif 'Christmas Day' in date_str: 
        year = date_str[-4:]
        return pd.to_datetime('{}-12-25'.format(year))
    elif 'Boxing Day' in date_str: 
        year = date_str[-4:]
        return pd.to_datetime('{}-12-26'.format(year))
    else: 
        return pd.to_datetime(date_str[4:], format='%d %b %Y')

df['date_tidied'] = df['date'].map(tidy_date)
```
